(1) In this assignment, you test whether the central limit theorem works. You generate 1000 variables with two normal distributions. You can determine the mean and standard deviation of these variables yourself. All you have to do is generate the first variable 50 times and averaged it each time. Generate the second variable 1000 times and averages this variable each time. Then plot the histogram of the averages of the two variables. Which of the variables has a mean distribution closer to the normal distribution? Do you think the Central Limit Theorem seems to have worked?

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(15,8))

populasyon1 = np.random.normal(15,0.1,1000)
populasyon2 = np.random.normal(15,0.1, 1000) 

print("populasyon1 ortalaması : ", populasyon1.mean())
print("populasyon2 ortalaması : ", populasyon2.mean())
print("populasyon1 standart sapması : ", populasyon1.std())
print("populasyon2 standart sapması : ", populasyon2.std())

orneklem1 = []
orneklem_ortalamalari1 = []
for atis_sayisi in range(1,50):
    atis = np.random.rand()
    orneklem1.append(atis)   
    orneklem_ortalamalari1.append(np.mean(orneklem1))

orneklem2 = []
orneklem_ortalamalari2 = []
for atis_sayisi in range(1,100):
    atis = np.random.rand()
    orneklem2.append(atis)   
    orneklem_ortalamalari2.append(np.mean(orneklem2))

def orneklem_ciz(örneklem_ortalaması,orneklem_sayisi, subplot_no):
    plt.subplot(subplot_no)
    plt.plot(örneklem_ortalaması[0:orneklem_sayisi])
    plt.title("Örneklem Sayısı = {}".format(orneklem_sayisi))
    plt.xlabel("Örneklem Sayısı")
    plt.ylabel("Örneklem Ortalaması")
    return plt
plt.figure(figsize=(15,3), dpi=100)
orneklem_ciz(orneklem_ortalamalari1, 50, 121)
orneklem_ciz(orneklem_ortalamalari2, 100, 122)
plt.show()
```
